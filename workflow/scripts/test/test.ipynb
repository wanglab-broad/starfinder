```python
sample_dict = {}
for current_sample in df['sample_id'].unique():
    fov_list = []

    current_start = df.loc[df['sample_id'] == current_sample, 'fov_start'].values[0]
    current_end = df.loc[df['sample_id'] == current_sample, 'fov_end'].values[0]

    for i in range(current_start, current_end+1):
        current_fov = "Position{i:03}".format(i=i)
        fov_list.append(current_fov)
    
    sample_dict[current_sample] = fov_list
```
